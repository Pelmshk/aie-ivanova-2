```python
import numpy as np
```

```python
# создание матрицы а 4х4 со случайными числами
a = np.array([[8, 3, 1, 4], [3, 7, 1, 2], [0, 9, 2, 10], [2, 5, 2, 9]])
a
```

```python
# размеры матрицы
a.shape
```

```python
# вывод типа данных элементов матрицы
a.dtype
```

```python
# создание единичной матрицы
eye = np.eye(4)
eye
```

```python
# поэлементное умножение матриц
a * eye
```

```python
# агрегация
print(a.max(), a.min(), a.mean())
```

```python
# вывод элемента матрицы по индексу
print(a[2][1])
```

```python
# срез
a[1:3]
```

```python
# срез строки матрицы
a[3][0:2]
```

```python
# Изменение формы
a.ravel()
```

```python
a.reshape(2, 8)
```

```python
# Объединение матриц
np.hstack((a, eye))
```
